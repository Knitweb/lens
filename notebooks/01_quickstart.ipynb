```python
from knitweb_lens import library, search

# Browse the built-in library
lib = library()
print(f"Library contains {len(lib)} circuits")
print()
for name, circ in list(lib.items())[:10]:
    print(f"  {circ.meta.name:<30} [{circ.meta.domain}]  {circ.meta.description[:50]}")
```

```python
# Load the Bell |Φ+⟩ state
bell = lib['bell_phi_plus']
print(f"Name   : {bell.meta.name}")
print(f"Qubits : {bell.meta.qubits}")
print(f"Tags   : {bell.meta.tags}")
print(f"CID    : {bell.cid}")
print()
print(bell.qasm)
```

```python
# Search circuits by keyword
results = search('grover')
print(f"Found {len(results)} Grover variants:")
for r in results:
    m = r['meta']
    print(f"  {m['name']:<30}  {m['description'][:60]}")
```

```python
# Import from QASM string directly
from knitweb_lens.adapters.qasm import from_qasm

my_qasm = '''
OPENQASM 2.0;
include "qelib1.inc";
qreg q[3];
creg c[3];
h q[0];
cx q[0],q[1];
cx q[1],q[2];
measure q[0] -> c[0];
measure q[1] -> c[1];
measure q[2] -> c[2];
'''

circ = from_qasm(my_qasm, name='my_ghz', domain='fundamental', tags=['ghz', 'custom'])
print(f"Created: {circ}")
print(f"CID    : {circ.cid}")
```

```python
# Store and retrieve
from knitweb_lens.store import Store

store = Store()  # local store at ~/.knitweb_lens/
cid = store.put(circ)
print(f"Stored with CID: {cid}")

retrieved = store.get(cid)
print(f"Retrieved: {retrieved.meta.name}")
assert retrieved.qasm == circ.qasm
print('Content-verified!')
```
